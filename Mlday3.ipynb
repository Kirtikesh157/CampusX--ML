```python
import  pandas  as pd
import  numpy as np 
import  matplotlib  as plt  
import  seaborn  as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size')
```

```python
titanic =pd.read_csv('titanic train.csv')
titanic.head()
```

```python
titanic.sample(5)
```

--- Boxplot

```python
sns.boxplot(x='Sex',y='Age',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',data=titanic, hue='Survived')
```

--- Barplot  

```python
sns.barplot(x='Pclass',y='Fare',data=titanic,)
```

```python
sns.barplot(x='Pclass',y='Fare',data=titanic,hue='Sex')
```

```python
sns.barplot(x='Pclass',y='Age',data=titanic,hue='Sex')
```

```python
sns.barplot(x='Pclass',y='Age',data=titanic)
```

--- Distplot


```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

--- heatmap

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']), annot=True, fmt='d')
```

```python
titanic.sample(3)
```

```python
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```

----clustermap 

```python
sns.clustermap(pd.crosstab(titanic['SibSp'], titanic['Survived']), annot=True, fmt='d')
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
sns.heatmap(flights.pivot_table(index='month',columns='year',values='passengers'))
```

```python
sns.clustermap(flights.pivot_table(index='month',columns='year',values='passengers'))
```

--- Lineplot  


```python
sns.lineplot(x='year',y='passengers',data=flights)
```

```python
sns.lineplot(x='year',y='passengers',data=flights,hue='month')
```
